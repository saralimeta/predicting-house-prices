### outliers

```python
def count_outliers_iqr(df):
    outliers_count = {}
    for column in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        num_outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
        outliers_count[column] = num_outliers
    return outliers_count

# Count outliers for all numerical columns in the dataset
outliers_count = count_outliers_iqr(data)

# Print the number of outliers for each column
for column, count in outliers_count.items():
    print(f"Number of outliers in column '{column}': {count}")
```

# PreProcessing

### Removing outliers

```python

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return cleaned_df

# Apply outlier removal to the 'BHK' column
cleaned_data = remove_outliers_iqr(data, 'BHK')
cleaned_data = remove_outliers_iqr(cleaned_data, 'Rent')
cleaned_data = remove_outliers_iqr(cleaned_data, 'Size')

```

## Feature Engineering and Selecting

### Encoding Values

```python
data = cleaned_data[['BHK', 'Size', 'City', 'Rent', 'Point of Contact', 'Bathroom','Furnishing Status', 'Tenant Preferred']]
```

```python
def one_hot_encode(data, column):
    encoded = pd.get_dummies(data[column], drop_first= True)
    data = data.drop(column, axis = 1)
    data = data.join(encoded)
    return data
```

```python
data = one_hot_encode(data, ['City', 'Point of Contact', 'Furnishing Status', 'Tenant Preferred'])
data
```
